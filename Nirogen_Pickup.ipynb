```python
import pandas as pd

# 1. Load your data (adjust paths as needed)
df1 = pd.read_csv("Heat_Information.csv", encoding='ISO-8859-1')

# 2. Clean up the original Heat No column
df1['Heat No'] = df1['Heat No'].astype(str).str.strip()

# 3. Define a function to expand entries like '199282/83/84'
def expand_heat_nos(heat_no):
    parts = heat_no.split('/')
    # If there's only one part, return it as-is
    if len(parts) == 1:
        return [heat_no]
    # Otherwise, use the first part as base, then append each suffix
    base = parts[0][:-2]
    return [parts[0]] + [base + p for p in parts[1:]]

# 4. Apply expansion and explode into individual rows
df1['Expanded Heat No'] = df1['Heat No'].apply(expand_heat_nos)
df1_expanded = df1.explode('Expanded Heat No').reset_index(drop=True)

# 5. Drop the old column and rename the expanded one
df1_expanded = (
    df1_expanded
    .drop(columns=['Heat No'])
    .rename(columns={'Expanded Heat No': 'Heat No'})
)

# 6. (Optional) Preview and save
print(df1_expanded['Heat No'].head(10))
df1_expanded.to_csv("Heat_Information_expanded.csv", index=False)
```

```python
df = pd.read_excel('/content/ouput.xlsx', sheet_name='Sheet1')
# Clean the data: drop rows where Sample_Id or NPP is missing and ensure Sample_Id is string
df_cleaned = df.dropna(subset=['Sample_Id', 'NPP']).copy()
df_cleaned['Sample_Id'] = df_cleaned['Sample_Id'].astype(str)

# Function to process each heat group
def process_heat_group(group):
    heat_no = group['Heat No'].iloc[0]

    # Step 1: Get highest L_ value
    l_samples = group[group['Sample_Id'].str.startswith('L')]
    l_max = l_samples.sort_values('Sample_Id').iloc[-1]['NPP'] if not l_samples.empty else None

    # Step 2: Choose X, Y, Z in order of priority
    ccm_group = group[group['Location'] == 'CCM']
    ccm_value = None
    for prefix in ['X', 'Y', 'Z']:
        sub = ccm_group[ccm_group['Sample_Id'].str.startswith(prefix)]
        if not sub.empty:
            ccm_value = sub.sort_values('Sample_Id').iloc[-1]['NPP']
            break

    return pd.Series({'Heat No': heat_no, 'L': l_max, 'CCM_Output': ccm_value})

# Apply the function grouped by Heat No
result_df = df_cleaned.groupby('Heat No').apply(process_heat_group).reset_index(drop=True)
result_df.head()
```
